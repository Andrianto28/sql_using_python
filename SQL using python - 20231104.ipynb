```python
# Query SELECT
query = """
SELECT *
    FROM batting
    ;
"""
```

```python
batting_observations = pd.read_sql(query, connection)
batting_observations
```

```python
# Select column
query = """
SELECT PlayerID, YearID
    FROM allstarfull
    ;
"""
select_columns = pd.read_sql(query, connection)
select_columns
```

```python
# Query Distinct (digunakan untuk menampilkan data unik untuk kolom yang dicari)
query = """
SELECT DISTINCT TeamID
    FROM allstarfull
    ;
"""
select_distinct = pd.read_sql(query, connection)
select_distinct
```

```python
# Query LIMIT (digunakan untuk menampilkan data sesuai limit (batas) yang diinginkan)
query = """
SELECT PlayerID, YearID,TeamID
    FROM allstarfull
    LIMIT 5
    ;
"""
select_limit = pd.read_sql(query, connection)
select_limit
```

```python
# Query Fully Qualified Names (FQN)
query = """
SELECT allstarfull.PlayerID
    FROM allstarfull
    LIMIT 5
    ;
"""
select_FQN = pd.read_sql(query, connection)
select_FQN
```

```python
# Query Order By (digunakan untuk melakukan sorting)
query = """
SELECT *
    FROM allstarfull
    ORDER BY playerID
    LIMIT 10
    ;
"""
select_order = pd.read_sql(query, connection)
select_order
```

```python
# Query WHERE (digunakan untuk mengambil data dengan kondisi)
query = """
SELECT *
    FROM allstarfull
    WHERE YearID = 2002
    ;
"""
select_where = pd.read_sql(query, connection)
select_where
```

```python
# Coba gabungkan syntax

query = """
SELECT *
    FROM allstarfull
    WHERE YearID = 2002
    ORDER BY PlayerID
    LIMIT 20
    ;
"""
select_combined = pd.read_sql(query, connection)
select_combined
```
